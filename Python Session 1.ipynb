Write a Python Program to Arithmetic Operations

```python
#Addition
num1 = float(input("Enter first number: "))
num2 = float(input("Enter second number: "))
Sum = num1 + num2
print(f"Sum: {num1} + {num2} = {Sum}")
```

```python
#Multiplication
Multiply = num1 * num2
print(Multiply)
```

Write a Python Program to Calculate Area of basic shapes

```python
#Area of Triangle
Base = float(input("Enter the base value: "))
Height = float(input("Enter the height value: "))
#Formula
Area = 0.5 * Base * Height
#Dispaly message
print(f"Area of Triangle is {Area}")
```

```python
#Area of Square
Side = float(input("Enter the value: "))
#Formula
Area = Side**2
#Display message
print(f"Area of Square is {Area}")
```

```python
#Area of Rectangle
Length = float(input("Enter the length: "))
Width = float(input("Enter the width: "))
#Formula
Area = Length * Width
#Display message
print(f"Area of Reactangle is {Area}")
```

Write a Python Program to convert Kilometers to miles

```python
Kilometers = float(input("Enter distance in kilometers: "))
#Conversion factor: 1 km = 0.621371 miles
conv_fact = 0.621371
#Formula
miles = Kilometers * conv_fact
print(f"{Kilometers} kilometers is equal to {miles} miles")
```

Write a Python Program to convert inch to centimeters

```python
Inches = float(input("Enter the Value: "))
#Conversion factor: 1 Inch = 2.54 cm
Conv_fact = 2.54
#Formula
Centimeters = Inches * Conv_fact
print(f"{Inches} inches is equal to {Centimeters} cm")
```
